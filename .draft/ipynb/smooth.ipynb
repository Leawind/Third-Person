```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def smoothFactor2Halflife(smoothFactor):
    return - 1 / np.log2(smoothFactor)

def halflife2SmoothFactor(halflife):
    return 2 ** (-1 / halflife)
```

```python
s = np.linspace(0, 1, 100)
h = smoothFactor2Halflife(s)
plt.figure(figsize=(10, 10))
plt.xlim(0, 1)
plt.plot(s, h)
```

```python
h = np.linspace(0, 1, 2048)
s = halflife2SmoothFactor(h)
plt.figure(figsize=(10, 10))
# plt.xlim(0, 0.5)
plt.plot(h, s)
```
