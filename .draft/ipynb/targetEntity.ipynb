```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def plot3D(x, y, func):
    X, Y = np.meshgrid(x, y)
    Z = func(X, Y)

    fig = plt.figure(figsize=(14, 14))
    ax = fig.add_subplot(111, projection='3d')

    surf = ax.plot_surface(X, Y, Z, cmap='viridis')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.view_init(elev=45, azim=45)
    plt.show()
```

```python
def func2(x, y):
    dist = np.sqrt(x**2 + y**2)
    angle = np.rad2deg(np.arctan2(x, y))
    angle = np.abs(angle)
    
    z = 1 / (dist**1e-3 * angle**7e-4)
    return z

SIZE = 128

x = np.linspace(-32, 32, SIZE)
y = np.linspace(0, 32, SIZE)
z = np.linspace(0, 32, SIZE)

plot3D(x, y, func2)
```

```python
def scatter3D(x, y, z, func):
    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(x, y, z, c=func(x,y,z), alpha=0.1, cmap='viridis', marker='.')  # 可以根据z值调整颜色

    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    
    ax.view_init(elev=36, azim=40 + 270)

    plt.show()
```

```python
SIZE = 50 **3
RANGE = 32
x = np.random.uniform(-RANGE, RANGE, SIZE)
y = np.random.uniform(0, RANGE, SIZE)
z = np.random.uniform(-RANGE, 0, SIZE)

def func3(x, y, z):
    dist = np.sqrt(x**2 + y**2 + z**2)
    angle = np.rad2deg(np.arctan2(np.sqrt(x**2 + z**2), y))
    angle = np.abs(angle)
    # z = dist**2 * angle**2.5
    z = dist * 1 + angle * 5
    return np.log(z)

scatter3D(x, y, z, func3)
```
