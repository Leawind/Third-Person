```python
import numpy as np
import matplotlib.pyplot as plt

def plot(x, y):
	plt.xlim(0, 1)
	plt.ylim(0, 1)
	plt.plot(x, y)
```

```python
x = np.arange(0, 1, 0.001)
```

```python
fc = x
plot(x, fc)
```

```python
fr_min = 0.3
fr = (1 - fr_min )* x + fr_min
plot(x, fr)
```

```python
C = 0.180 * 1.1
fr = (x-C) / (1-C)
plot(x, fr)
```
