```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# 创建一些示例数据
# 假设我们有三个坐标轴的数据
n_points = 1000
np.random.seed(0)
x = np.random.randn(n_points)
y = np.random.randn(n_points)
z = np.random.randn(n_points)


# 创建一个新的图形窗口，并添加一个3D坐标轴
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# 使用scatter函数绘制三维散点图
ax.scatter(x, y, z, c=z, cmap='viridis', marker='o')  # 可以根据z值调整颜色

# 设置坐标轴标签
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# 显示图形
plt.show()
```
